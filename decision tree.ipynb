```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pdf=pd.read_csv("salaries.csv")
```

```python
pdf
```

```python
input=pdf[["company","job","degree"]]
```

```python
input
```

```python
y=pdf["salary_more_then_100k"]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
input["com_n"]=le_company.fit_transform(input["company"])
```

```python
input["job_n"]=le_job.fit_transform(input.job)
```

```python
input["degree_n"]=le_degree.fit_transform(input.degree)
input
```

```python
final=input.drop(["company","job","degree"],axis=1)
```

```python
final
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
model.fit(final,y)
```

```python
model.score(final,y)
```

```python
model.predict([[2,2,0]])
```
